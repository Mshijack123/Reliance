```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from datetime import datetime
df = pd.read_csv('/kaggle/input/reliance/RELIANCE.NS.csv')
df.head()

sns.set(rc={'figure.figsize' :(10,5)})
```

```python
df.isnull().sum()
df['Date'] = pd.to_datetime(df.Date)
df = df.set_index("Date")
df.head()
##year23=df[(df['Date'] >= '2023-01-01') & (df['Date'] < '2023-04-01') ]

##year23.isnull().sum()
##year23.head()
```

```python



sns.lineplot(x=df.index , y = df['Open'] , label='Open')
```

```python
df.plot(y=['High','Close','Open'] , title = 'Reliance')
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3, figsize = (15,10))
df.groupby(df.index.day).mean().plot(y = 'Volume' , ax=ax1 ,xlabel='day')
df.groupby(df.index.month).mean().plot(y = 'Volume' ,ax=ax2 , xlabel='month')
df.groupby(df.index.year).mean().plot(y = 'Volume' ,ax=ax3 , xlabel='year')
```

```python
a = df.sort_values(by = 'High', ascending= False).head(5)
a


sns.barplot(data = a , x = a.index.day , y= a.High )
```

```python
df.head()
year2023 = df[df.index.year == 2023]

year2023
```

```python
year2023.isnull().sum()
year2023.dropna(how='all')
```

```python
sns.lineplot(data = year2023 , x=year2023.index , y='Close')
```

```python
sns.barplot(data = year2023 , x=year2023.index.month , y='Close')
```

```python
short_rolling = year2023.rolling(window=5).mean()
##short_rolling.head()

long_rolling = year2023.rolling(window=20).mean()
##short_rolling.head()
```

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(year2023.index , year2023.Close , label='Price')
ax.plot(short_rolling.index , short_rolling.Close ,color='red', label='SMA 5 days')

ax.plot(long_rolling.index , long_rolling.Close , label='SMA 20 days')
```
